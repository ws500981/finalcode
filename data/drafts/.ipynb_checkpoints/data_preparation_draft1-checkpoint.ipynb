```python
from Bio import SeqIO
import Bio
with open("SRR10971019_subreads.fastq") as handle:
    records = list(SeqIO.parse(handle, "fastq"))
    print(records[0].id)  # first record
    print(records[-1].id)  # last record
```

```python
print(type(records)) # records is a list containing all the records
```

```python
print(type(records[0])) # each record is of type Bio.SeqRecord.SeqRecord
```

```python
print(type(records[-1].id)) # the record id is a string
int(records[-1].id[12:]) # this is the id number for each record, there are a total of 95514 records
```

```python
#for each record, convert the sequence to string. Method 1 (preferred because there is no title and no ending '\n')
for i in range(0,int(records[-1].id[12:])): 
    first_record = records[i]
    print(type(first_record.seq))
    strrep = str(first_record.seq)
    print("STRREP", type(strrep), len(strrep))
    print(strrep)
    break
```

```python
#for each record, convert the sequence to string. Method 2 (not preferred due to the title and ending '\n')
for i in range(0,int(records[-1].id[12:])): 
    first_record = records[i]
    print(type(first_record))
    tab = Bio.SeqIO.TabIO.as_tab(first_record)
    print("TAB", type(tab), len(tab[14:]))
    print(tab[14:])
    break
```

```python
tab2 = tab[14:15761] # removed the title and the ending '\n' for sequence generated with method 1
tab2 == strrep # now the two results are equal (note that even when the ending index in 
            # the first line is 15762 they will also be equal because tab2 will be one character (the '\n') longer than strrep)
print(tab[15762]=='\n')
print(tab[15761])
print(strrep[15747])
```

```python
#now use either tab2 or strrep as a string sequence to be processed to have the required format

#format1: non-overlapping (if (len_after-remainder)/7 == (len_before-remainder)/6 then the splitting is correct)
#use the same method to insert '\n' such that each line contains 65 6-mers
# [TODO] save the string as tsv
n = 0
print("before:", tab2, len(tab2))
while n<=len(tab2) and len(tab2)-n>7:
    if n == 0:
        n = 6
        tab2 = tab2[:n] + "\t" + tab2[n:]
    else:
        n = n + 7
        tab2 = tab2[:n] + "\t" + tab2[n:]
print("after: ", tab2, len(tab2))

n = 0
print("before:", tab2, len(tab2))
while n<=len(tab2) and len(tab2)-n>456:
    if n == 0:
        n = 455
        tab2 = tab2[:n] + "\n" + tab2[n:]
    else:
        n = n + 456
        tab2 = tab2[:n] + "\n" + tab2[n:]
print("after: ", tab2, len(tab2))
```
