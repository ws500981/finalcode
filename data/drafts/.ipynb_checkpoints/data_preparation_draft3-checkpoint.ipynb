```python
from __future__ import print_function
from Bio import SeqIO
import Bio
import csv
with open("SRR10971019_subreads.fastq") as handle:
    records = list(SeqIO.parse(handle, "fastq"))
```

```python
#USED, CAN GET CORRECT RESULT, non-overlapping
#for each record, convert the sequence to string. Method 1 (preferred because there is no title and no ending '\n')
with open("output_nonoverlapping.tsv", "w", newline = '') as f:
    for i in range(0,int(records[-1].id[12:])): 
        first_record = records[i]
        #print(type(first_record.seq)) # type is <class 'Bio.Seq.Seq'>
        strrep = str(first_record.seq)
        print("STRREP", len(strrep)) # type(strrep) = <class 'str'>
        #print(strrep) #print out the sequence
        
        strrep = strrep[:len(strrep)-len(strrep)%6]

        #format1: non-overlapping (if (len_after-remainder)/7 == (len_before-remainder)/6 then the splitting is correct)
        #use the same method to insert '\n' such that each line contains 65 6-mers
        # [TODO] save the string as tsv
        n = 0
        while n<=len(strrep) and len(strrep)-n>7:
            if n == 0:
                n = 6
                strrep = strrep[:n] + "\t" + strrep[n:]
                #print("n=",n)
            else:
                n = n + 7
                strrep = strrep[:n] + "\t" + strrep[n:]
                #print("n=",n)
        print("after adding tabs: ", len(strrep))
        arr = strrep.split("\t")
        #print(arr)
        tsv_output = csv.writer(f, delimiter=' ')
        flag = True
        for p in range(0,len(arr)+1,65):
            arr1 = []
            for q in range(0,64):
                try:
                    arr1.append(arr[p+q])
                except:
                    flag = False
            if flag:
                #print("before: ",arr1[63])
                arr1[63] = arr1[63] + '\t' + arr[p+q+1]
                #print("after: ",arr1[63])
                #print(arr1)
                tsv_output.writerow(str(arr1))
        break
```

```python
#USED, CAN GET CORRECT RESULT, sliding window by kmers
#for each record, convert the sequence to string. Method 1 (preferred because there is no title and no ending '\n')
with open("output_slidingbykmer2.tsv", "w", newline = '') as f:
    for i in range(0,int(records[-1].id[12:])): 
        first_record = records[i]
        #print(type(first_record.seq)) # type is <class 'Bio.Seq.Seq'>
        strrep = str(first_record.seq)
        print("STRREP", len(strrep)) # type(strrep) = <class 'str'>
        #print(strrep) #print out the sequence
        
        strrep = strrep[:len(strrep)-len(strrep)%6]
        n = 0
        while n<=len(strrep) and len(strrep)-n>7:
            if n == 0:
                n = 6
                strrep = strrep[:n] + "\t" + strrep[n:]
                #print("n=",n)
            else:
                n = n + 7
                strrep = strrep[:n] + "\t" + strrep[n:]
                #print("n=",n)
        print("after adding tabs: ", len(strrep))
        arr = strrep.split("\t")
        #print(arr)
        tsv_output = csv.writer(f, delimiter=' ')
        flag = True
        for p in range(0,len(arr)+1):
            arr1 = []
            for q in range(0,64):
                try:
                    arr1.append(arr[p+q])
                except:
                    flag = False
            if flag:
                print("before: ",arr1[63])
                arr1[63] = arr1[63] + '\t' + arr[p+q+1]
                print("after: ",arr1[63])
                print(arr1)
                tsv_output.writerow(arr1)
        break
```

```python
#USED, CAN GET CORRECT RESULT, sliding window by base
import sys
with open("output_slidingbybase.tsv", "w", newline = '') as f:
    #for i in range(0,int(records[-1].id[12:])): 
    first_record = records[0] # REMEMBER TO CHANGE THIS BACK TO i
    #print(type(first_record.seq)) # type is <class 'Bio.Seq.Seq'>
    strrep = str(first_record.seq)
    print("STRREP", len(strrep)) # type(strrep) = <class 'str'>
    #print(strrep) #print out the sequence

    strrep = strrep[:len(strrep)-len(strrep)%70]
    n = 0
    while n<=len(strrep) and len(strrep)-n>71:
        if n == 0:
            n = 70
            strrep = strrep[:n] + "\t" + strrep[n:]
            #print("n=",n)
        else:
            n = n + 71
            strrep = strrep[:n] + "\t" + strrep[n:]
            #print("n=",n)
    print("after adding tabs: ", len(strrep))
    arr = strrep.split("\t")
    #print(arr)

    separator = ''
    for element in arr:
        print(len(element))
        sentence = ""
        base = 0
        while base<=64:
            if base == 0:
                tempList = []
                for q in range(0,6):
                    tempList.append(element[q])
                word = separator.join(tempList)
                #print(word, base, sentence)
                base = 1
                sentence = word
            elif base==64:
                tempList = []
                for q in range(0,6):
                    tempList.append(element[base+q])
                word = separator.join(tempList)
                #print(word, base, sentence)
                base += 1
                sentence = sentence + "\t" + word
            else:
                tempList = []
                for q in range(0,6):
                    tempList.append(element[base+q])
                word = separator.join(tempList)
                #print(word, base, sentence)
                base += 1
                sentence = sentence + " " + word
        print("%s" % (sentence), file=f)
        print(sentence)
```

```python
import copy
import csv
import json
import logging
import os
class InputExample(object):
    def __init__(self, guid, text_a, text_b=None, label=None):
        self.guid = guid
        self.text_a = text_a
        self.text_b = text_b
        self.label = label

    def __repr__(self):
        return str(self.to_json_string())

    def to_dict(self):
        """Serializes this instance to a Python dictionary."""
        output = copy.deepcopy(self.__dict__)
        return output

    def to_json_string(self):
        """Serializes this instance to a JSON string."""
        return json.dumps(self.to_dict(), indent=2, sort_keys=True) + "\n"
```
